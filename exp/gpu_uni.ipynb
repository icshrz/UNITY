```python
import yaml
import matplotlib.pyplot as plt

# 从 YAML 文件读取数据
input_file = "gpu_uni2.yaml"

with open(input_file, "r") as file:
    data = yaml.load(file, Loader=yaml.FullLoader)

# 设置时间戳的最小值和最大值
min_timestamp = 0.0  # 最小时间戳值
max_timestamp = 15.0  # 最大时间戳值

# 提取时间戳和 GPU 利用率，且只选择时间戳在给定范围内的数据
timestamps = [float(item['timestamp']) for item in data if min_timestamp <= float(item['timestamp']) <= max_timestamp]
gpu_usages = [item['gpu_usage'] for item in data if min_timestamp <= float(item['timestamp']) <= max_timestamp]

# 绘制折线图，调整线条粗细和颜色
plt.figure(figsize=(8, 2.5))
plt.plot(timestamps, gpu_usages, label="GPU Utilization", color='#5ea69c', linewidth=1)  # 浅绿色线条
plt.xlabel("Time (seconds)")
plt.ylabel("GPU Utilization (%)")
plt.title("GPU Utilization")
plt.tight_layout()
plt.savefig("exp61.pdf", bbox_inches='tight')

# 显示图表
plt.show()
```
