```python
import yaml
import matplotlib.pyplot as plt
import numpy as np

# 从 YAML 文件读取数据（图1）
input_file_1 = "modified_gpu_usage_data.yaml"

with open(input_file_1, "r") as file:
    data_1 = yaml.load(file, Loader=yaml.FullLoader)

# 设置时间戳的最小值和最大值
min_timestamp = 0.0  # 最小时间戳值
max_timestamp = 15.0  # 最大时间戳值

# 提取时间戳和 GPU 利用率，且只选择时间戳在给定范围内的数据
timestamps_1 = [float(item['timestamp']) for item in data_1 if min_timestamp <= float(item['timestamp']) <= max_timestamp]
gpu_usages_1 = [item['gpu_usage'] for item in data_1 if min_timestamp <= float(item['timestamp']) <= max_timestamp]

# 从 YAML 文件读取数据（图2）
input_file_2 = "gpu_uni2.yaml"

with open(input_file_2, "r") as file:
    data_2 = yaml.load(file, Loader=yaml.FullLoader)

# 提取时间戳和 GPU 利用率，且只选择时间戳在给定范围内的数据
timestamps_2 = [float(item['timestamp']) for item in data_2 if min_timestamp <= float(item['timestamp']) <= max_timestamp]
gpu_usages_2 = [item['gpu_usage'] for item in data_2 if min_timestamp <= float(item['timestamp']) <= max_timestamp]

input_file_3 = "io.yaml"

with open(input_file_3, "r") as file:
    io_data = yaml.load(file, Loader=yaml.FullLoader)

# 获取数据
gdelt_data = io_data['GDELT']

# 提取数据
# 更新后的数据
tglt_io = gdelt_data['TGN']['TGL']['I/O']
tglt_memcpy = gdelt_data['TGN']['TGL']['Memcpy']
tgnunity_io = gdelt_data['TGN']['UNITY']['I/O']
tgatgl_io = gdelt_data['TGAT']['TGL']['I/O']
tgatgl_memcpy = gdelt_data['TGAT']['TGL']['Memcpy']
tgatunity_io = gdelt_data['TGAT']['UNITY']['I/O']

# 绘制折线图和柱状图的子图，并调整子图的大小
fig = plt.figure(figsize=(8, 3))

# 使用 gridspec 来定义子图布局
gs = fig.add_gridspec(1, 2, width_ratios=[3, 1])  # 设置第一个子图占宽度的3倍，第二个子图占1倍

# 创建折线图子图
ax1 = fig.add_subplot(gs[0])
ax1.plot(timestamps_1, gpu_usages_1, label="MSPipe", color='#5ea69c', linewidth=1)
ax1.plot(timestamps_2, gpu_usages_2, label="UNITY", color='#ff6347', linewidth=1)
ax1.set_xlabel("Time (seconds)")
ax1.set_ylabel("GPU Utilization (%)")
ax1.set_title("GPU Utilization Comparison")
ax1.legend(loc='upper right')

# 创建柱状图子图
ax2 = fig.add_subplot(gs[1])
labels = ['TGN', 'TGAT']
bar_width = 0.3
index = np.arange(len(labels))

# 绘制堆叠柱状图（TGL部分）
ax2.bar(index - bar_width / 2, [tglt_io, tgatgl_io], bar_width, label='TGL I/O', color='#cccccc', edgecolor='black')
ax2.bar(index - bar_width / 2, [tglt_memcpy, tgatgl_memcpy], bar_width, label='Memcpy', color='white', edgecolor='black', bottom=[tglt_io, tgatgl_io])

# 绘制其他柱状图（UNITY部分）
ax2.bar(index + bar_width / 2, [tgnunity_io, tgatunity_io], bar_width, label='UNITY I/O', color='black', edgecolor='black')

ax2.set_ylabel("I/O Latency(ms)")
ax2.set_xticks(index)
ax2.set_xticklabels(labels)
ax2.legend(loc='upper center', ncol=1, bbox_to_anchor=(0.32, 1.03), frameon=False)

# 调整布局
plt.tight_layout()

# 保存图表
plt.savefig("exp6.pdf", bbox_inches='tight')

# 显示图表
plt.show()
```
